```python
# Shuffle the dataset for equal distribution
data= data.sample(frac=1).reset_index(drop=True)
data.head(10)

# Split the data
y=data['Result']
X=data.drop('Result',axis = 1)
```

# DECISION TREE CLASSIFIER

```python
# import packages
from sklearn.model_selection import train_test_split
# Split the data into testing and training data in the ratio 80:20

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size= 0.2, random_state = 2) # testing set = 20%
X_train.shape, X_test.shape
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion='entropy',max_depth=5)
tree=tree.fit(X_train, y_train)
```

```python
y_pred_test = tree.predict(X_test)
y_pred_train = tree.predict(X_train)
```

```python
# Model Evaluation

# Find Accuracy  for training and testing data
#import packages
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
acc_train = accuracy_score(y_train,y_pred_train)
acc_test = accuracy_score(y_test,y_pred_test)

# Print accuracy for training and testing data

print("Accuracy of Training Data: ",acc_train)
print("Accuracy of Testing Data:",acc_test)
```

```python
# Confusion Matrix

def CMatrix(CM, labels=['real','phis']):
    data=pd.DataFrame(data=CM, index=labels, columns=labels)
    data.index.name='True'
    data.columns.name='Prediction'
    data.loc['Total']=data.sum()
    data['Total']=data.sum(axis=1)
    return data
CM = confusion_matrix(y_pred_test,y_test)
CMatrix(CM)
```

```python
precision= precision_score(y_pred_test,y_test)
recall = recall_score(y_pred_test, y_test)
```

```python
print("Precision of the model: ",precision)
print("Recall of the model: ", recall)

f1=f1_score(y_pred_test,y_test, average='weighted')
print("F1 Score of the the model: ", f1)
```

# RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(max_depth=6)
forest.fit(X_train,y_train)
```

```python
y_for_train = forest.predict(X_train)
y_for_test = forest.predict(X_test)
```

```python
acc_for_train = accuracy_score(y_for_train,y_train)
acc_for_test = accuracy_score(y_for_test,y_test)
```

```python
print("Accuracy of Training model:", acc_for_train)
print("Accuracy of Testing model:", acc_for_test)
```

```python
# Model Evaluation 

CM=confusion_matrix(y_for_test,y_test)
CMatrix(CM)
```

```python
precision= precision_score(y_for_test,y_test)
recall= recall_score(y_for_test,y_test)
f1=f1_score(y_for_test,y_test)
```

```python
print("Precision of the model : ", precision)
print("Recall of the model: ", recall)
print("F1 Score of the model: ", f1)
```

```python
# Compare
ML_Model=['Decision Tree','Random Forest']
acc_training=[acc_train,acc_for_train]
acc_testing=[acc_test,acc_for_test]

results = pd.DataFrame({'ML_Model':ML_Model,
                       'Train Accuracy': acc_training,
                       'Test Accuracy': acc_testing})
results
```
